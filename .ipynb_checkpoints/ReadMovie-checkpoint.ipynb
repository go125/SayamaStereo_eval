# Extract Frames

```python
import cv2
import os
```

```python
def extractFrames(pathIn, pathOut ,option="None"):
    num=0
    if not os.path.exists(pathOut):
        os.mkdir(pathOut)

    cap = cv2.VideoCapture(pathIn)
    count = 0

    while (cap.isOpened()):

        ret, frame = cap.read()

        if ret == True:
            num+=1
            if num%3==0:
                init_height, init_width = frame.shape[:2]
                #print(init_height)
                #print(init_width)
                HEIGHT=init_height//3
                WIDTH=init_width
                if option=="top":
                    frame =frame[0:HEIGHT,0:WIDTH] 
                if option=="middle":
                    frame =frame[HEIGHT:(HEIGHT*2),0:WIDTH] 
                if option=="bottom":
                    frame =frame[(HEIGHT*2):(HEIGHT*3),0:WIDTH] 
                cv2.imwrite(os.path.join(pathOut, "frame_{:06d}.jpg".format(count)), frame)
                count += 1
            if num%300==0:
                print(count)
        else:
            break

    cap.release()
    #cv2.destroyAllWindows()
```

```python
def main1():
    extractFrames('/home/ubuntu/Sayama/data/StereoVideo/V2-mv-20200716103312-ulrg.avi', '/home/ubuntu/Sayama/video1top/',"top")
```
